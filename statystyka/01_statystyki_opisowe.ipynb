## 1. Statystki opisowe

### Populacja vs prbka



Wykonujemy pomiar (co w statystyce nazywamy czsto robieniem eksperymentu). Zanim zaczniemy analizowa otrzymany wynik, kluczowe jest rozrnienie czy pomiar dotyczy caej **populacji, ang. population**, czy tylko jej czci (**prbki, ang. sample**). Wane jest te czy na podstawie otrzymanego wyniku chcemy wnioskowa na temat caej populacji, czy tylko samej prbki. To zagadnienie  jest kamie wgielnym caej dziedziny i powodem jej istnienia.

Na ile pomiar wycznie czci zbioru uprawnia nas do wycigania wnioskw dotyczcych caego zbioru. Z jak du pewnoci moemy wypowiada si na temat caego zbioru posiadajc informacje dotyczce wycznie czci tego zbioru (prbki).

Dlaczego moe nas to interesowa? Poniewa bardzo czsto zebranie informacji na temat caej populacji jest zbyt zasoboerne. Czsto w ogle niemoliwe. Z drugiej strony, stosunkowo niskim kosztem potrafimy zebra informacje o duym poziomie wiarygodnoci. Ten temat stanowi jedno z podstawowych zagadnie optymalizacyjnych w dzisiejszym wiecie.



### Parametry vs statystyki

Liczebno populacji oznaczamy symbolem **$N$**, liczebno prbki symbolem **$n$**.

Po zebraniu danych wykonujemy obliczenia, eby otrzyma jakie miarodajne informacje na temat naszych danych. W przypadku populacji wynik oblicze nazywamy **parametrem** populacji, w przypadku prbki nazywamy **statystyk**.

Przykad: Przeprowadzamy ankiet wrd studentw Uniwerystetu Warszawskiego.

Wniosek: Populacja jest trudna do obserwacji. Testy statystyczne powstay po to, ebymy mogli pracowa na niekompletnych danych. Mc na ich podstawie wyciga prawidowe wnioski. Przecie najczciej pracujemy z prbk, maym fragmentem zbioru wszystkich moliwych informacji.

```python
#populacja i prbka
import numpy as np

N = 1000 # liczebno populacji
n = 10  # liczebno prbki (sample)

# Tworzymy "populacj" (liczby cakowite od 1 do 1000)
populacja = list(range(1, N+1))

# Losujemy prbk z populacji
sample_size = n  # Rozmiar prby
przyklad_proby = np.random.choice(populacja, size=sample_size, replace=False)

# Wyliczamy i wywietlamy redni populacji i prbki
srednia_populacji = np.mean(populacja)  # parametr
srednia_proby = np.mean(przyklad_proby)  # statystyka

print(f"Populacja: {populacja}")
print()
print(f"Przykadowa prba: {przyklad_proby}")
print()
print(f"rednia populacji: {srednia_populacji}")
print(f"rednia prby: {srednia_proby}")
```
