Dane ze wzgldu na poziom pomiaru mona podzieli na dane:
- jakociowe (aka kwalitatywne, ang. qualitative)
    - nominalne (ang. nominal)
    - porzdkowe (ang. ordinal)
- ilociowe (aka kwantytatywne, ang. quantitative)
    - interwaowe (ang. interval)
    - ilorazowa (ang. ratio)

#### Dane jakociowe (*ang. qualitative*)

Dane jakociowe to dane, ktre opisuj cechy lub kategorie i nie maj wartoci liczbowej. Ich celem jest klasyfikacja obiektw lub zjawisk na podstawie jakociowych atrybutw. Danych jakociowych nie mona uporzdkowa liczbowo ani wykorzysta do przeprowadzania operacji matematycznych.

Dane jakociowe moemy podzieli na:
- dane nominalne (*ang. nominal*) - dane bez porzdku (np. kolory, pe, marka samochodu, gatunek zwierzcia).
- dane porzdkowe (*ang. ordinal*) -  dane, ktre mona uporzdkowa, ale rnice midzy nimi nie maj okrelonej wartoci (np. stopnie w szkole, pory roku, opinia zbierana w ankiecie)

#### Dane ilociowe (*ang. quantitative*) 

Dane ilociowe to dane, ktre opisuj ilo i maj wartoci liczbowe. Te dane umoliwiaj wykonywanie operacji matematycznych, takich jak dodawanie, odejmowanie, czy rednia arytmetyczna. Mog by mierzone lub zliczane i s kluczowe w analizach statystycznych.

Dane ilociowe moemy podzieli na:
- dane interwaowe (*ang. interval*) - dane liczbowe nie posiadajce prawdziwego zera (aka rzeczywiste zero, absolutny punkt zerowy). Zero jest tu tylko punktem na skali, a nie brakiem wartoci. Moliwe s operacje dodawania i odejmowania, ale nie mona sensownie porwnywa stosunkw midzy wartociami; nie ma sensu mwienie, e 10 stopni jest dwa razy cieplejsze ni 5 stopni. Dlaczego ?



Patrzc na stopnie w skali Celciusa moglibymy powiedzie, e dzisiaj jest dwa razy cieplej ni wczoraj, ale patrzc na stopnie w skali Farenheita ju nie. Dlaczego tak jest? Poniewa O w skali Celciusa i 0 w skali Farenheita nie s prawdziwymi zerami. Te skale zostay sztucznie stworzone przez czowieka. Przykady danych interwaowych: temperatura w skali C i F, lata w kalendarzu)
- dane ilorazowe (*ang. ratio*) - dane liczbowe posiadajce prawdziwe zero. Moliwe s wszystkie operacje matematyczne, w tym sensowne jest porwnywanie stosunkw midzy wartociami; sens ma powiedzenie e co way dwa razy wicej (np. waga, wzrost, dugo, temperatura mierzona w Kelvinach 0K=-213.15C=-459.67F). Wikszo rzeczy, ktre obserwujemy w wiecie s wanie tego typu. Nazwa (ratio) pochodzi std, e mog one przedstawia relacj midzy rnymi rzeczami. Na przykad, jeli mam 2 jabka, a ty masz 6, to ty masz ich 3 razy wicej ni ja. Jak to obliczyem? Stosunek 6 do 2 wynosi 3. Inne przykady to liczba przedmiotw w ogle, odlego czy czas.

Na koniec, liczby takie jak 2, 3, 10, 10,5 czy $\pi$ mog by zarwno danymi interwaowymi, jak i ilorazowymi. Trzeba by ostronym i uwzgldnia kontekst, w ktrym si je stosuje.
