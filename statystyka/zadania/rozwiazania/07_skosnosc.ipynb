```python
data2 = [
    586,
    760,
    495,
    678,
    559,
    415,
    370,
    659,
    119,
    288,
    241,
    787,
    522,
    207,
    160,
    526,
    656,
    848,
    720,
    676,
    581,
    929,
    653,
    661,
    770,
    800,
    529,
    975,
    995,
    947,
]
```

```python
# Zadanie 1
import pandas as pd

# Tworzenie serii danych
df1 = pd.Series(data1)

# Obliczanie współczynnika asymetrii (skewness) dla obu zbiorów
skew1 = df1.skew()

# Wyniki
print(f"Asymetria w zbiorze danych 1: {skew1}")
```

```python
# Zadanie 2
import matplotlib.pyplot as plt
import seaborn as sns  # oczywiście można byłoby to zrobić tylko z matplotlibem, 
# ale dla urozmaicenia popatrzmy na seaborn

# Tworzenie histogramów
plt.figure(figsize=(12, 6))

# Histogram dla data1
plt.subplot(1, 2, 1)
sns.histplot(df1, kde=True, bins=10, color='blue')  # kde=True nanosi ciągłą aproksymację histogramu (tą niebieską krzywą)
plt.title("Histogram dla danych 1")
plt.xlabel("Wartości")
plt.ylabel("Częstość")

# Wykres pudełkowy dla data1
plt.subplot(1, 2, 2)
sns.boxplot(data=df1, color='blue')
plt.title("Wykres pudełkowy dla danych 1")

plt.tight_layout()
plt.show()
```

```python
# Zadanie 2

df2 = pd.Series(data2)
skew2 = df2.skew()

print(f"Asymetria w zbiorze danych 2: {skew2}")
```

```python
# Histogram dla data2

# Tworzenie histogramów
plt.figure(figsize=(12, 6))

# Histogram dla data1
plt.subplot(1, 2, 1)
sns.histplot(df2, kde=True, bins=10, color='blue')
plt.title("Histogram dla danych 2")
plt.xlabel("Wartości")
plt.ylabel("Częstość")

# Wykres pudełkowy dla data1
plt.subplot(1, 2, 2)
sns.boxplot(data=df2, color='blue')
plt.title("Wykres pudełkowy dla danych 2")

plt.tight_layout()
plt.show()
```

#### Interpretacja

**Histogram** pozwala zobaczyć rozkład danych i sprawdzić, czy są one asymetryczne.

**Wykres pudełkowy** pomaga dostrzec medianę, kwartyle oraz wartości odstające (outliers). Jeśli wykres pudełkowy ma długi ogon po jednej stronie, oznacza to asymetrię.
