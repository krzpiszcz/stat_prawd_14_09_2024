### 4. Tabela krzyżowa

Tło: \
Dysponujesz danymi dotyczącymi zatrudnienia w kraju X. Poproszono Cię o przygotowanie tabeli krzyżowej, która przedstawi te dane.

60% osób w wieku od 18 do 25 lat jest zatrudnionych\
85% osób w wieku od 25 do 35 lat jest zatrudnionych\
5% osób w wieku od 35 do 45 lat jest bezrobotnych\
3% osób w wieku od 45 do 55 lat jest bezrobotnych\
3% osób w wieku od 55 do 65 lat jest bezrobotnych\
Wszyscy powyżej 65. roku życia są zatrudnieni.

Uwaga: \
Definicja bezrobotnych to osoby bez pracy, ale aktywnie jej poszukujące. To dlatego wszyscy powyżej 65. roku życia są zatrudnieni.

Zadanie 1: \
Utwórz tabelę krzyżową podsumowującą przedstawione dane.

Zadanie 2: \
Utwórz wykres słupkowy side-by-side, aby wizualnie wzmocnić swoje podsumowanie.

```python
# Zadanie 1
import pandas as pd

# Definiujemy dane
age_groups = ['18-25', '25-35', '35-45', '45-55', '55-65', '65+']
employed_percent = [60, 85, 95, 97, 97, 100]  # Procent zatrudnionych
unemployed_percent = [100 - employed for employed in employed_percent]  # Procent bezrobotnych

# Tworzymy DataFrame
data = pd.DataFrame({
    'Zatrudnieni (%)': employed_percent,
    'Bezrobotni (%)': unemployed_percent
}, index=age_groups,)
data  # to jest nasza tablica krzyżowa
```

```python
# Zadanie 2
import matplotlib.pyplot as plt
import numpy as np

# Ustawienie szerokości słupków
bar_width = 0.35
index = np.arange(len(age_groups))

# Tworzymy wykres
plt.figure(figsize=(10, 6))

# Słupki dla zatrudnionych
plt.bar(index, data['Zatrudnieni (%)'], bar_width, label='Zatrudnieni', color='green')

# Słupki dla bezrobotnych
plt.bar(index + bar_width, data['Bezrobotni (%)'], bar_width, label='Bezrobotni', color='red')

# Dodajemy etykiety, tytuł i legendę
plt.xlabel('Grupa wiekowa')
plt.ylabel('Procent (%)')
plt.title('Procent zatrudnionych i bezrobotnych w zależności od grupy wiekowej')
plt.xticks(index + bar_width / 2, age_groups)  # Ustawiamy etykiety na osi X
plt.legend()

# Wyświetlamy wykres
plt.tight_layout()
plt.show()
```
