```python
# Zadanie 3
import numpy as np 

# Obliczamy zakres danych i dokładną szerokość przedziału
min_value = min(data)
max_value = max(data)
range_value = max_value - min_value
interval_width = range_value / 6  # Dokładna szerokość przedziału, bez zaokrąglania

# Tworzymy przedziały
bins = []
for i in np.arange(min_value, max_value, interval_width):
    bins.append((i, i + interval_width))

# Obliczamy częstość absolutną
frequency = []
for idx, (bin_start, bin_end) in enumerate(bins):
    if idx == 0:
        count = sum(1 for x in data if bin_start <= x <= bin_end)
    else:
        count = sum(1 for x in data if bin_start < x <= bin_end)
    frequency.append(count)

# Obliczamy częstość względną
total_count = len(data)
relative_frequency = [count / total_count for count in frequency]

# Tworzymy tabelę rozkładu
table = pd.DataFrame({
    'Przedziały': [f'{bin_start:.2f} - {bin_end:.2f}' for bin_start, bin_end in bins],
    'Częstość absolutna': frequency,
    'Częstość względna': relative_frequency
})

# Wyświetlamy tabelę
table
```
