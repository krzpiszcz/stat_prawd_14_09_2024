### 10. Kowariancja

Tło: \
Dysponujesz danymi kilku uczniów dotyczącymi wyników z egzaminu z czytania i z egzaminu z pisania.

Zadanie 1: \
Określ, czy jest to próba, czy cała populacja.

Zadanie 2: \
Oblicz kowariancję dla tych dwóch zbiorów danych.

Zadanie 3: \
Narysuj dane na wykresie punktowym i, korzystając z wcześniejszej wiedzy, skomentuj, czy istnieje zauważalny związek między tymi dwiema zmiennymi.

Zadanie 4: \
Oblicz współczynnik korelacji Pearsona dla tych dwóch zbiorów danych.

Zadanie 5: \
Przedyskutuj siłę korelacji między tymi dwoma zbiorami danych.

```python
writing_scores = [
    344,
    383,
    611,
    713,
    536
]
```

```python
reading_scores = [
    378,
    349,
    503,
    719,
    503
]
```

#### Zadanie 1

Na podstawie dostępnych danych (tylko 5 wyników), możemy założyć, że mamy do czynienia z próbą z populacji, a nie z całą populacją. Więc musimy obliczyć kowariancję i korelację zakładając, że to próbka, a nie pełna populacja.

```python
# Zadanie 2
import pandas as pd
import matplotlib.pyplot as plt

# Tworzymy DataFrame
data = pd.DataFrame({
    'Writing': writing_scores,
    'Reading': reading_scores
})

# Obliczanie kowariancji
covariance = data.cov().loc['Writing', 'Reading']
print(f"Kowariancja: {covariance}")
```

```python
# Zadanie 3 

# Wykres punktowy
plt.figure(figsize=(8, 6))
plt.scatter(data['Writing'], data['Reading'], color='blue', label='Dane')
plt.title('Wykres punktowy: Wyniki z pisania vs wyniki z czytania')
plt.xlabel('Wyniki z pisania')
plt.ylabel('Wyniki z czytania')
plt.grid(True)
plt.legend()
plt.show()
```

Na podstawie bardzo małej próby widzimy wyraźną korelację. Być może, gdyby próba była większa korelacja okazałaby się słabsza.

```python
# Zadanie 4

# Obliczanie współczynnika korelacji Pearsona
correlation = data.corr().loc['Writing', 'Reading']
print(f"Współczynnik korelacji Pearsona: {correlation}\n")

# Interpretacja
if correlation > 0.8:
    print("Istnieje silna dodatnia korelacja między wynikami z pisania i czytania.")
elif correlation < -0.8:
    print("Istnieje silna ujemna korelacja między wynikami z pisania i czytania.")
else:
    print("Korelacja jest słaba lub umiarkowana.")
```

#### Zadanie 5

Jeśli współczynnik korelacji jest bliski 1, możemy powiedzieć, że istnieje silna dodatnia korelacja między wynikami z egzaminów z pisania i czytania. Oznacza to, że uczniowie, którzy dobrze napisali egzamin, zazwyczaj dobrze poradzili sobie również z egzaminem z czytania.

Jeśli współczynnik korelacji jest bliski 0, to może oznaczać brak związku między wynikami z tych dwóch egzaminów.


U nas współczynnik korelacji wynosi 0.92. Wysoka dodatnia korelacja między wynikami z pisania i czytania sugeruje, że uczniowie, którzy osiągają dobre wyniki w jednym obszarze, zazwyczaj osiągają podobne wyniki w drugim.
