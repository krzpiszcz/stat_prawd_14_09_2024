## Rachunek prawdopodobieństwa

Dlaczego omawiamy rachunek prawdopodobieństwa?

Ponieważ podstawowa wiedza z rachunku prawdopodobieństwa będzie nam potrzebna do tzw. wnioskowania statystycznego (**ang. inferential statistic**). Wnioskowanie statystyczne jest zbiorem narzędzi oraz technik zajmujących się wnioskowaniem na temat wartości parametrów populacji na podstawie fragmentu tej populacji (tzw. próbki, ang. sample).

Zacznijmy od pojęcia dystrybucji.

### Dystrybucja (ang. distribution)

Czym jest dystrybucja?

W statystyce, gdy używamy terminu dystrybucja (aka rozkład), zazwyczaj mamy na myśli rozkład prawdopodobieństwa. Dobrym przykładem są rozkład normalny, rozkład dwumianowy i rozkład jednostajny.

**Rozkład to funkcja, która pokazuje możliwe wartości dla zmiennej i częstość ich występowania.**

Najlepiej popatrzmy na jakiś prosty przykład.

Rzucanie sześcienną kostką do gry - to nasz eksperyment. 

Jakie są możliwe wyniki? 1, 2, 3, 4 ,5 ,6

Jakie jest prawdopobieństwo, że wynikiem eksperymentu będzie 1? \
$\frac{1}{6}$ 

Jakie jest prawdopobieństwo, że wypadnie (wynikiem eksperymentu będzie) 2? \
$\frac{1}{6}$

A jakie jest prawdopobieństwo, że wypdanie 7? \
0

A w ogólności ?

```python
import pandas as pd

result = pd.DataFrame({
    "wynik": [1, 2, 3, 4, 5, 6],
    "prawdopodobienstwo": [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
})
result
```

W ogólności rozkład prawdopodobieństwa "składa się" nie tylko z wyników możliwych do zaobserwowania, ale ze wszystkich wyników. Warunkiem na to, żeby funkcja była rozkładem prawdopodobieństwa jest to, że po zsumowaniu prawdopodobieństw wszystkich wyników dostaniem 1 (100%).

W ten sposób otrzymaliśmy rozkład prawdopodobieństwa rzutów sześcienną kostką do gry. Nanieśmy wynik na wykres.

```python
result.plot(kind='bar', x='wynik', y='prawdopodobienstwo', legend=False)
```

Taki rozkład nazywamy rozkładem jednostajnym aka równomiernym (**ang. uniform distribution**). Jest to jednocześnie rozkład dyskretny. Czyli:

**zmienna losowa opisująca wynik eksperymentu polegającego na rzucie sześciościenną kostką do gry podlega dyskretnemu rozkładowi jednostajnemu**. 

Pamiętaj! Rozkład to nie wykres. Rozkład stanowią wartości prawdopodobieństwa otrzymania poszczególnych wyników. Wykres jest wyłącznie reprezentacją graficzną tego rozkładu.

Uwaga! Jeżeli na kursie analizy matematycznej zetknąłeś się z dystrybucjami (nazywanymi również funkcjami uogólnionymi lub funkcjami Schwartza), to nie są to dystrybucje w rozumieniu statystycznym. Nie należy ich ze sobą mylić, a jedyne co łączy obie koncepcje to nazwa.

### A. Rozkład normalny

Istnieje wiele popularnych rozkładów, ale my skupimy się głównie na dwóch: normalnym i rozkładzie t-studenta z kilku powodów, w tym:
- tymi dwoma rozkładami można przybliżyć ogromną liczbę zmiennych
- rozkład próby o wystarczająco duże liczebności może być przybliżony rozkładem normalnym

To co to jest rozkład normalny ?

Rozkład normalny, zwany również rozkładem Gaussa, rozkładem Gaussa-Laplace'a lub krzywą dzwonową, to jedno z najważniejszych pojęć w statystyce i teorii prawdopodobieństwa. Jest to rozkład ciągły, który często pojawia się w kontekście różnych zjawisk naturalnych, takich jak wyniki pomiarów, rozkład wzrostu ludzi, błędy pomiarowe czy też wyniki testów. Rozkład ten jest symetryczny i ma charakterystyczny kształt dzwonu.

Wyraża się wzorem:

$$f(x)=\frac{1}{\sigma\sqrt{2\pi}} e^{(-\frac{1}{2}(\frac{x-\mu}{\sigma})^2)}$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parametry rozkładu normalnego
mu = 0  # średnia
sigma = 1  # odchylenie standardowe
rozmiar = 10000  # liczba próbek

# Generowanie danych z rozkładu normalnego
data = np.random.normal(mu, sigma, rozmiar)

# Tworzymy wykres histogramu z gęstością prawdopodobieństwa
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')

# Dodanie krzywej dzwonowej na podstawie wzoru analitycznego
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
plt.plot(x, y, 'k', linewidth=2)

# Ustawienia wykresu
plt.title("Histogram i krzywa dzwonowa - Rozkład normalny")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.grid(True)
plt.show()
```

Rozkład jest symetryczny w związku z tym średnia, mediana i dominanta rozkładu są sobie równe.

Rozkład normalny często zapisuje się za pomocą formuły: 

$$N \sim (\mu, \sigma^2)$$

która oznacza, że jest to rozkład o średniej $\mu$ i wariancji $\sigma^2$.

W naszym przykładzie, rozkład ma średnią 0. Utrzymując stałe odchylenie standardowe (lub mówiąc statystycznym językiem, kontrolując odchylenie standardowe), niższa średnia spowoduje, że rozkład zachowa ten sam kształt, ale znajdzie się bardziej na lewo. 

Niech $\mu=-1$ (patrz zielony wykres).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parametry rozkładu normalnego
mu = 0  # średnia
sigma = 1  # odchylenie standardowe
rozmiar = 10000  # liczba próbek

# Generowanie danych z rozkładu normalnego
data = np.random.normal(mu, sigma, rozmiar)

# Tworzymy wykres histogramu z gęstością prawdopodobieństwa
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')

# Dodanie krzywej dzwonowej na podstawie wzoru analitycznego
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
mu2 = -1
y2 = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu2) / sigma) ** 2)
plt.plot(x, y, 'k', linewidth=2)
plt.plot(x, y2, 'g', linewidth=2)

# Ustawienia wykresu
plt.title("Histogram i krzywa dzwonowa - Rozkład normalny")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.grid(True)
plt.show()
```

Na tej samej zasadzie zwiększenie średniej spowoduje przesunięcie całego wykresu (bez zmiany jego kształtu) na prawo.

Niech $\mu=1$ (patrz wykres czerwony).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parametry rozkładu normalnego
mu = 0  # średnia
sigma = 1  # odchylenie standardowe
rozmiar = 10000  # liczba próbek

# Generowanie danych z rozkładu normalnego
data = np.random.normal(mu, sigma, rozmiar)

# Tworzymy wykres histogramu z gęstością prawdopodobieństwa
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')

# Dodanie krzywej dzwonowej na podstawie wzoru analitycznego
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
mu2 = 1
y2 = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu2) / sigma) ** 2)

plt.plot(x, y, 'k', linewidth=2)
plt.plot(x, y2, 'r', linewidth=2)

# Ustawienia wykresu
plt.title("Histogram i krzywa dzwonowa - Rozkład normalny")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parametry rozkładu normalnego
mu = 0  # średnia
sigma = 1  # odchylenie standardowe
rozmiar = 10000  # liczba próbek

# Generowanie danych z rozkładu normalnego
data = np.random.normal(mu, sigma, rozmiar)

# Tworzymy wykres histogramu z gęstością prawdopodobieństwa
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0, color='blue')

# Dodanie krzywej dzwonowej na podstawie wzoru analitycznego
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
mu2 = -1
y2 = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu2) / sigma) ** 2)
mu3 = 1
y3 = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu3) / sigma) ** 2)

plt.plot(x, y, 'k', linewidth=2)
plt.plot(x, y2, 'g', linewidth=2)
plt.plot(x, y3, 'r', linewidth=2)

plt.axvline(0, ymax=0.94, color='k', ls='--')
plt.axvline(-1, ymax=0.94, color='g', ls='--')
plt.axvline(1, ymax=0.94, color='r', ls='--')

plt.annotate(text=r"$\mu=0$", xy=(0, 0), xytext=(-0.2, -0.04))
plt.annotate(text=r"$\mu=-1$", xy=(0, 0), xytext=(-1.4, -0.02), color='g')
plt.annotate(text=r"$\mu=1$", xy=(0, 0), xytext=(0.8, -0.02), color='r')

# Ustawienia wykresu
plt.title("Krzywa dzwonowa - Rozkład normalny")
plt.ylabel("P(x)")
plt.grid(True)
plt.show()
```

Na identycznej zasadzie możemy kontrolując średnią, zmienić odchylenie standardowe i zobaczyć, co się stanie. Tym razem wykres nie przesunie się, lecz zmieni kształt.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parametry rozkładu normalnego
mu = 0  # średnia
sigma = 1  # odchylenie standardowe
rozmiar = 10000  # liczba próbek

# Generowanie danych z rozkładu normalnego
data = np.random.normal(mu, sigma, rozmiar)

# Tworzymy wykres histogramu z gęstością prawdopodobieństwa
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0, color='blue')

# Dodanie krzywej dzwonowej na podstawie wzoru analitycznego
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
sigma2 = 0.5
y2 = (1 / (sigma2 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma2) ** 2)
sigma3 = 2
y3 = (1 / (sigma3 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma3) ** 2)

plt.plot(x, y, 'k', linewidth=2)
plt.plot(x, y2, 'g', linewidth=2)
plt.plot(x, y3, 'r', linewidth=2)

plt.annotate(text=r"$\sigma=0$", xy=(0, 0), xytext=(-1.8, 0.22))
plt.annotate(text=r"$\sigma=0.5$", xy=(0, 0), xytext=(-1.4, 0.54), color='g')
plt.annotate(text=r"$\sigma=2$", xy=(0, 0), xytext=(-3, 0.12), color='r')

# Ustawienia wykresu
plt.title("Krzywa dzwonowa - Rozkład normalny")
plt.ylabel("P(x)")
plt.grid(True)
plt.show()
```
